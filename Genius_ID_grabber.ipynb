```python
import requests
from Genius_keys import client_access_token
req={'access_token':client_access_token}
```

```python
def find_artist_id(r, artist_name):
    
    r = r.json()
    artist_name=artist_name.lower() #lower names to remove possiblilties of non-matches. Some may still fall through the cracks.
    
    #Iterate through each hit in our search.
    for hit in r["response"]["hits"]:
        
        result_name = hit["result"]["primary_artist"]["name"].lower() #get the name of primary artist for each hit. Avoid songs they are featured in
        
        if result_name == artist_name:
            
            artist_id = hit["result"]["primary_artist"]["id"]
            
            #Once found, end our search
            return artist_id 
        
    return 'No ID found' #If none of the hits return an ID for our artist, return
```

```python
def get_artist_songs(artist_id,params):
    # initialize variables & a list.
    params = dict(params) #make a copy just in case we don't want to change our original request
    current_page = 1
    next_page = True
    songs = []

    # main loop
    while next_page:

        path = r'http://api.genius.com/artists/'+str(artist_id)+'/songs?page='+str(current_page)
        params['page']=current_page
        
        data = requests.get(path,params=params)
        data=data.json()
        page_songs = data['response']['songs']

        if page_songs:
            # add all the songs of current page,
            # and increment current_page value for next loop.
            songs += page_songs
            current_page += 1
        else:
            # if page_songs is empty, quit.
            next_page = False

    # get all the song titles,ids,paths excluding not-primary-artist songs.
    songs = [[song['title'],song["id"],song['path']] for song in songs
             if song["primary_artist"]["id"] == artist_id]

    return songs
```

Let's get the Genius ID for every artist in our file that we can!

```python
url=r'http://api.genius.com/search?q='
artist_id_list =[]
with open('hip_hop_artists.txt','r',encoding='utf-8') as infile:
    for name in infile:
        
        name=name.split('\n')[0] #remove carriage returns
        search_name = name.replace(' ','%20') #Convert whitespace to %20 for our search
        print(name)
        
        new_url = url+search_name
        r = requests.get(new_url,params=req)
        artist_id = find_artist_id(r,name)
        if artist_id != 'No ID found':
            artist_id_list.append([name, artist_id])
        
```

```python
headers = ['artist','genius_id']
with open('hip_hop_artist_ids.txt','w',encoding='utf-8') as outfile:
    outfile.write('\t'.join(headers) + '\n')
    
    for pair in artist_id_list:
        name =pair[0]
        artist_id = str(pair[1])
        outfile.write('\t'.join([name,artist_id]) + '\n')
```

Now let's build a file of all songs for each individual artist/group. We create a function to get the completed progress before we start requesting from the API. This is here just in case you want to interrupt the loop and not run this for 100000000 straight hours.

```python
def Find_progress(output_file):
    completed_artists=[]
    try:
        with open(output_file,'r',encoding='utf-8') as infile:
            for line in infile:
                artist = line.split('\t')[3].strip() #artist name
                completed_artists.append(artist) #keep track of who we have completed
    except:
        pass
    
    if len(completed_artists)==0:
        mode='w' #if we have no progress, we will be writing
    else:
        mode='a' #if we have progress, we will be appending
    return completed_artists,mode
```

```python
output_filename='song_ids_full.txt'
completed_artists,mode=Find_progress(output_filename)
  
            
with open('hip_hop_artist_ids.txt','r',encoding='utf-8') as infile:
    with open(output_filename,mode,encoding='utf-8') as outfile:
          for line in infile:        
            artist = line.split('\t')[0].strip()
            if (artist in completed_artists) or (artist=='artist'):
                continue      
            artist_id = int(line.split('\t')[1].strip())
            all_songs = get_artist_songs(artist_id,req) #get data
            all_songs = [song+[artist,artist_id] for song in all_songs] #include artist name, id
            
            #don't keep instrumentals or remixes
            all_songs = [song for song in all_songs if ('Remix' or 'Instrumental' or 'remix' or'instrumental') not in song[0]]
            
            #convert all to strings for writing
            for song in all_songs:
                song[1]=str(song[1])
                song[4]=str(song[4])
                
            #write out data
            for song in all_songs:
                outfile.write('\t'.join(song) + '\n')
            
```
